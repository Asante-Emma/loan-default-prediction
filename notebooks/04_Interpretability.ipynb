```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load model and test set
model = joblib.load('../models/final_random_forest_model.joblib')
X_test = pd.read_csv('../data/processed/X_test.csv')
y_test = pd.read_csv('../data/processed/y_test.csv')
```

## Feature Importance (Global Explainability)

```python
feature_importances = pd.Series(model.feature_importances_, index=X_test.columns)
top_features = feature_importances.sort_values(ascending=False).head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_features, y=top_features.index, palette='viridis', hue=top_features.index, legend=False)
plt.title('Top 15 Feature Importances (Random Forest)')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('../reports/figures/random_forest_feature_importance.png')
plt.show()
```

- **Loan percent income**, which measures how much of the applicants income the loan represents, is by far the most influential feature.

- **Interest rate**, **loan amount**, and **personal income** are also strong indicators of default likelihood.

- Borrowers with longer employment length and higher credit history length are generally less likely to default.

- Categorical features such as **loan intent**, **grade**, and **home ownership** play a secondary role but still contribute meaningfully.
